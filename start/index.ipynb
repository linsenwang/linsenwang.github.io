```python
import json
```

```python
with open('./infinityBackup-2024-5-12.infinity','r') as file:
    raw = json.load(file)
    print(raw)
```

```python
for i in raw:
    print(i)
```

```python
data = raw['data']
for i in data:
    print(i)
```

```python
sites = data['site']['sites'][0]
for i in sites:
    print(i)
```

```python
with open('img_src.html','w') as f:
    for i in sites:
        try:
            target = i['target']
            try:
                bgColor = i['bgColor']
            except KeyError:
                pass
            f.write(i['bgImage']+'\n')
        except KeyError:
            pass
```

```python
with open('infinity.html','w') as f:
    f.write('<link rel="stylesheet" href="infinity.css">')
    f.write('<meta name="viewport" content="width=device-width,initial-scale=1.0,maximum-scale=1.0,minimum-scale=1.0,user-scalable=no">')
    f.write('<body><div id=main>')
    for i in sites:
        try:
            target = i['target']
            f.write('<a href="'+ target + '">')
            f.write('<img ')
            try:
                bgColor = i['bgColor']
                if bgColor != '':
                    f.write('style="background-color: '+bgColor+';"')
            except KeyError:
                pass
            f.write('src="'+i['bgImage']+'">')
            f.write('\n')
        except KeyError:
            pass
    f.write('</div></body>')
```
